Keras implementation of GLO (Optimizing the Latent Space of Generative Networks)
https://arxiv.org/pdf/1707.05776.pdf

```python
from keras.models import Model, Sequential
from keras.layers import Dense, Embedding, Flatten, Dropout
from keras.constraints import UnitNorm
from keras.datasets import mnist
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
EMBEDDING_SIZE = 32
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = (x_train / 255).reshape(-1, 28*28)
```

# Building model

```python
model = Sequential()
model.add(Embedding(x_train.shape[0], EMBEDDING_SIZE, embeddings_constraint=UnitNorm(axis=1), input_length=1))
model.add(Flatten())
model.add(Dense(128, activation='relu', input_shape=(EMBEDDING_SIZE,)))
model.add(Dropout(0.3))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(28*28, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(np.arange(0, x_train.shape[0]).reshape(-1,1), x_train, batch_size=256, epochs=300)
```

# Recovering training data

```python
recovered = model.predict(np.arange(0, x_train.shape[0]).reshape(-1,1))
```

```python
idx = 1

plt.imshow(x_train[idx].reshape((28,28)), 'gray')
plt.show()
plt.imshow(recovered[idx].reshape((28,28)), 'gray')
plt.show()
```

# UMAP projection

```python
from umap import UMAP
projector = UMAP()

embeddings = model.layers[0].get_weights()[0]
embeddings_projected = projector.fit_transform(embeddings)
```

```python
plt.scatter(embeddings_projected[:,0], embeddings_projected[:,1], c=y_train, alpha=0.1)
```

# Sampling from multivariate normal distribution

```python
sampler_model = Sequential(model.layers[2:])
sampled_data = sampler_model.predict(np.random.normal(size=(32, EMBEDDING_SIZE)))

plt.imshow(sampled_data[0].reshape((28,28)), 'gray')
```

# Sampling from fitted gaussian mixture

```python
from sklearn.mixture import BayesianGaussianMixture
embeddings = model.layers[0].get_weights()[0]

bgm = BayesianGaussianMixture(n_components=64)
bgm.fit(embeddings)
```

```python
sampled_data = sampler_model.predict(bgm.sample(n_samples=32)[0])
plt.imshow(sampled_data[0].reshape((28,28)), 'gray')
```
